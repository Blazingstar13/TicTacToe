```
class TicTacToe:    
    #вызывается от объекта класса
    #инициализия переменных (начальное поле, текущее состояние и значок)
    #ничего не выводит    
    def __init__(self):        
        self.field = [1,2,3,4,5,6,7,8,9]
        self.currentState = self.field
        self.currentBadge = "O"         #изначально ходит О
    #------------------------

    #вызывается от объекта класса
    #изменение значка (X, O)
    #ничего не выводит
    def changeBadge(self):
        if self.currentBadge == "O":
            self.currentBadge = "X"
        else:
            self.currentBadge = "O"
    #------------------------        
    
    #вызывается от объекта класса
    #нарисовать поле
    #ничего не выводит     
    def drawField(self):
        print("—"*13)         #для красивой таблицы                                                           
        for i in range(3):
            print("|",self.field[0+i*3],"|",self.field[1+i*3],"|",self.field[2+i*3],"|")
            print("—"*13)      
    #------------------------

    #вызывается от объекта класса. Передается координата клетки
    #функция перехода состояния
    #выводит новое состояние или -1, если ячейка занята
    def newState(self, step):            # step - ячейка, в которую вписать значок        
        if(str(self.field[step-1]) not in "XO"):    #проверка, что ячейка не занята
             self.field[step-1] = self.currentBadge              
             self.changeBadge()                      #изменение значка после хода
             return self.field                       #возвращает новое состояние
        else:
             return "-1"                             #возвращает -1, если ячейка занята
    #------------------------

    

    #вызывается от объекта класса
    #конечные состояния
    #выводит значок победителя или False, если никто еще не победил
    def endState(self):
        win = ((0,1,2), (3,4,5), (6,7,8), (0,3,6), (1,4,7), (2,5,8), (0,4,8), (2,4,6)) #выигрышные ячейки
        for i in win:
            if self.field[i[0]] == self.field[i[1]] == self.field[i[2]]:  #проверка на равность всех трех выигрышных позиций
                return self.field[i[0]]           #возвращает значок победителя ( Х или О )
        return False
    #------------------------

    #вызывается от объекта класса
    #начало игры
    #ничего не выводит                    
    def play(self):        
        print("********   TicTacToe   ********")  
        counter = 1                             #счетчик шагов
        win = False                             #условие для выхода из первого цикла 
        while not win:                          #пока кто-то не выиграет        
            self.drawField()                    #нарисовать поле игры
            valid = False                       #условие для выхода из второго цикла
            while not valid:         
                command, step = input("In which cell to put " + self.currentBadge +"? ").split() #команда (grab) для хода и step - ячейка
                if command == "grab":
                    step = int(step)
                    if step >= 1 and step <= 9:                   #проверка на вхождение в диапазон                        
                        if self.newState(int(step)) == "-1":      #если быля ошибка
                            print("Oops, the cell is occupied!" )
                        else:
                            valid = True                          #выход из второго цикла
                    else:
                        print("Error, the entered number is not in the range from 1 to 9")
                else:
                    print("The command is wrong!")
                
            counter += 1                        #увеличение счетчика при ходе 
            if(counter > 4):                    #до 5 хода никто выиграть не может
                tmp = self.endState()           #проверка на победу
                if(tmp):
                    print(tmp, "won!")
                    win = True                  #выход из 1 цикла
                    break
            if(counter == 10):                  #было 9 ходов и никто не выиграл
                print("Draw!")
                break
        self.drawField()    
       #------------------------

#начало игры
game = TicTacToe()      #объект класса
game.play()             #начать игры
#------------------------
```

Автомат имеет начальное состояние поля(незаполненый список (от 1 до 9)), текущее поле, текущий значок(изначально О), множество конечных состояний(множество выигрышных позиций). Игра начинается с вызова метода play(). Пользователь вводит команду grab и координату клетки. Идет переход состояния. После 5 хода идет проверка на конечное состояние. Если после 9 переходов победителя не было выявлено, то будет ничья.
